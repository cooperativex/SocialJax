# Advanced Configuration in SocialJax

This tutorial covers advanced configuration options for fine-tuning your multi-agent RL experiments.

## Overview

SocialJax provides a flexible configuration system:

1. **YAML Configuration Files**: Store experiment settings in files
2. **Configuration Dataclasses**: Type-safe configuration objects
3. **Configuration Presets**: Pre-defined configurations for common scenarios
4. **CLI Overrides**: Override settings from command line

## 1. Import Required Components

```python
from socialjax.config import (
    ConfigManager,
    TrainingConfig,
    NetworkConfig,
    AlgorithmConfig,
    EnvironmentConfig,
    create_default_config,
)
from socialjax.training import Trainer
```

## 2. Configuration Dataclasses

```python
# TrainingConfig for training hyperparameters
training_config = TrainingConfig(
    total_timesteps=1_000_000,
    num_envs=8,
    num_steps=128,
    learning_rate=0.0003,
    gamma=0.99,
    gae_lambda=0.95,
    clip_eps=0.2,
    entropy_coef=0.01,
    value_loss_coef=0.5,
    max_grad_norm=0.5,
)

print("TrainingConfig:")
print(f"  Total timesteps: {training_config.total_timesteps}")
print(f"  Num envs: {training_config.num_envs}")
print(f"  Learning rate: {training_config.learning_rate}")
```

## 3. NetworkConfig

```python
# NetworkConfig for network architecture
network_config = NetworkConfig(
    network_type="cnn_small",
    hidden_size=128,
    num_layers=3,
    activation="relu",
)

print("NetworkConfig:")
print(f"  Network type: {network_config.network_type}")
print(f"  Hidden size: {network_config.hidden_size}")
print(f"  Activation: {network_config.activation}")
```

## 4. AlgorithmConfig

```python
# AlgorithmConfig combines training and network config
algorithm_config = AlgorithmConfig(
    name="ippo",
    training=training_config,
    network=network_config,
)

print("AlgorithmConfig:")
print(f"  Algorithm name: {algorithm_config.name}")
```

## 5. EnvironmentConfig

```python
# EnvironmentConfig for environment settings
env_config = EnvironmentConfig(
    name="coin_game",
    num_agents=5,
    max_steps=250,
)

print("EnvironmentConfig:")
print(f"  Environment: {env_config.name}")
print(f"  Num agents: {env_config.num_agents}")
print(f"  Max steps: {env_config.max_steps}")
```

## 6. Using ConfigManager

```python
# Create a ConfigManager
config_manager = ConfigManager()

# Create default config
default_config = create_default_config()
print("Default config created!")
print(f"Algorithm: {default_config.algorithm.name}")
```

## 7. Loading Configuration from YAML

```python
print("Loading config from YAML:")
print("")
print("# socialjax/config/presets/algorithms/ippo.yaml")
print("""algorithm:
  name: ippo
  training:
    learning_rate: 0.0003
    gamma: 0.99
    gae_lambda: 0.95
    clip_eps: 0.2
    entropy_coef: 0.01
  network:
    network_type: cnn_small
    hidden_size: 128
""")
```

```python
print("Loading preset from file:")
print("")
print("from socialjax.config import ConfigManager")
print("")
print("manager = ConfigManager()")
print("config = manager.load_from_file('socialjax/config/presets/algorithms/ippo.yaml')")
```

## 8. Merging Configurations

```python
print("Merging configurations:")
print("")
print("# Base config")
print("base = {")
print("    'algorithm': {'name': 'ippo'},")
print("    'training': {'learning_rate': 0.0003},")
print("}")
print("")
print("# Override config")
print("override = {")
print("    'training': {'learning_rate': 0.001},  # Higher LR")
print("}")
print("")
print("# Merged result")
print("# merged['training']['learning_rate'] == 0.001")
```

## 9. Configuration Presets

```python
import os

presets_dir = "socialjax/config/presets"
print("Available preset files:")
print("")

if os.path.exists(presets_dir):
    for root, dirs, files in os.walk(presets_dir):
        level = root.replace(presets_dir, '').count(os.sep)
        indent = ' ' * 2 * level
        print(f'{indent}{os.path.basename(root)}/')
        subindent = ' ' * 2 * (level + 1)
        for file in files:
            print(f'{subindent}{file}')
else:
    print("(Presets directory not found)")
```

## 10. Using Configuration with Trainer

```python
print("Using config with Trainer:")
print("")
print("# Method 1: Pass config dict")
print("trainer = Trainer(")
print("    algorithm='ippo',")
print("    env='coin_game',")
print("    config={")
print("        'total_timesteps': 1000000,")
print("        'learning_rate': 0.0003,")
print("        'num_envs': 8,")
print("    }")
print(")")
print("")
print("# Method 2: Load from file")
print("trainer = Trainer(")
print("    algorithm='ippo',")
print("    env='coin_game',")
print("    config_path='path/to/config.yaml',")
print(")")
```
